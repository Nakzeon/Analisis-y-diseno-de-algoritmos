Tienes un pipeline que recibe una lista de registros de transacciones. Cada registro forma un diccionario con valores transaction_id y una amount. Debido a problemas en el proceso, pueden aparecer IDs duplicados en la lista.

Escribe una función que reciba la lista y regrese una  una nueva lista sin duplicados, manteniendo solamente la  primera aparición de cada transacción.






<br>
Lenguaje R

```python
quitar_duplicados <- function(transacciones) {
  ids_recorrer <- c()
  limpio <- list()

  for(trs in transacciones){
    tx_id <- trs$transaction_id

    if(!(tx_id %in% ids_recorrer)){
      ids_recorrer <- c(ids_recorrer, tx_id)
      limpio <- c(limpio, list(trs))
    }
  }

  return(limpio)
}

transacciones <- list(
  list(transaction_id = "T1", amount = 100),
  list(transaction_id = "T2", amount = 200),
  list(transaction_id = "T1", amount = 100),  # duplicate
  list(transaction_id = "T3", amount = 300),
  list(transaction_id = "T2", amount = 999)   # duplicado, difefrent amount
)

output <- quitar_duplicados(transacciones)

for(t in output){
  cat("ID:", t$transaction_id, "| Amount:", t$amount, "\n")
}
```

```python
def quitar_duplicados(transacciones):
    ids_recorrer = set()
    limpio = []

    for trs in transacciones:
        tx_id = trs["transaction_id"]

        if tx_id not in ids_recorrer:
            ids_recorrer.add(tx_id)
            limpio.append(trs)

    return limpio

transacciones = [
    {"transaction_id": "T1", "amount": 100},
    {"transaction_id": "T2", "amount": 200},
    {"transaction_id": "T1", "amount": 100},  # duplicate
    {"transaction_id": "T3", "amount": 300},
    {"transaction_id": "T2", "amount": 999}   # duplicado, different amount
]

output = quitar_duplicados(transacciones)

for t in output:
    print(f"ID: {t['transaction_id']} | Amount: {t['amount']}")
```
